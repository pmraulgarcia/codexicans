Now we need to import the **libraries** that will be used in the program.
*   pandas
*   yfinance
*   matplotlib

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

**We need to define the stock(s) that we want to review.**
In this case Netflix.

```python
stock='NFLX'
```

**We create a yfinance object including the following parameters:**

*   stock
*   start date
*   end date

In this way, we define the stock that we want to review and the period of the query.

```python
stocks = yf.download(stock, start ="2010-01-01", end="2022-07-08")
```

**We use pandas library to store the values that we obtain from Yahoo Finance site**

```python
df = pd.DataFrame(stocks)
```

**We list the first 5 record of the query to validate the start date.**

```python
df.head()
```

**We list the last 5 record of the query to validate the start date.**

```python
df.tail()
```

To get additional information of an specific stock, we use the ticker module.

Wikipedia: A **ticker symbol or stock symbol** is an abbreviation used to uniquely identify publicly traded shares of a particular stock on a particular stock market.

We can compare the following information with the oficial information published in the official site for investors. [Go](https://ir.netflix.net/ir-overview/profile/default.aspx)

```python
nflx = yf.Ticker(stock)
```

**show earnings (Quaterly)**

Access just to the **closing** stock values.

```python
data = stocks.loc[:,"Close"].copy()
data
```

We use the plot function to **create a graph fo the historical values **of the selected stock, Netflix in this case.

```python
data.plot(figsize = (17, 8),fontsize =18)
plt.style.use("seaborn")
plt.title('Netflix',fontsize=18, color='blue')
plt.xlabel('Years',fontsize=14, color='blue')
plt.ylabel('Price',fontsize=14, color='blue')
plt.show()
```
